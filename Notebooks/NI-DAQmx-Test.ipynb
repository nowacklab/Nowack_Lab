```python
from Nowack_Lab.Instruments.nidaq import NIDAQ
```

```python
import matplotlib.pyplot as plt
```

```python
daq = NIDAQ()
daq.inputs = {
    'signal generator': 'ai1',
}
```

```python
samples = daq.monitor(
    chan_in = ['signal generator'],
    duration = 3e-3,#s
    sample_rate = 1e6,#Hz
)

plt.plot(samples['t'], samples['signal generator'])
plt.xlabel('Time (s)')
plt.ylabel('Signal generator (V)')
plt.show()
```
