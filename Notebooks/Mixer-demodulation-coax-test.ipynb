```python
import Nowack_Lab
import numpy as np
import matplotlib.pyplot as plt

from Nowack_Lab.Instruments import nidaq, e8241a

rf = e8241a.E8241a()
rf.write('*RST')
rf.write('*CLS')
rf.outputting = False
rf.amplitude = 10.0 - 40.0 + 6.0 - 3.0 #dBm
rf.frequency = 2e9#Hz

daq = nidaq.NIDAQ()
daq.inputs = {
    'IF': 'ai1',
}

rf.amplitude = 10.0 - 40.0 + 6.0 - 3.0 #dBm

def frequency_sample(f):
    rf.frequency = f
    samples = daq.monitor(
        chan_in = ['IF'],
        duration = 1e-3,#s
        sample_rate = 1e6,#Hz
    )
    return np.mean(samples['IF'])
```

For 10 dBm at LO, need -30 dBm from splitter and thus -24 dBm into splitter. Initially, extra -3 dBm for safety.

---

Initial testing: **June 10, 2022** to **June 14, 2022**

```python
frequencies = np.arange(2e9, 3e9, 1e6)
ifs = [frequency_sample(f) for f in frequencies]

plt.plot(frequencies, ifs)
plt.xlabel('Frequency (Hz)')
plt.ylabel('IF (V)')
plt.title('E8241A RF power: %s dBm (LO: +34 dB)' %rf.amplitude)
plt.show()
```

```python
np.mean(ifs), (np.max(ifs) - np.min(ifs))
```
